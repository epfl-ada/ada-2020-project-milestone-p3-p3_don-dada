```python
from uk_covid19 import Cov19API
import pandas as pd
```

```python

london_only = [
    'areaType=ltla',
]

cases_and_deaths = {
    "date": "date",
    "areaName": "areaName",
    "areaCode": "areaCode",
    "newCasesBySpecimenDate": "newCasesBySpecimenDate",
    "newCasesByPublishDate":"newCasesByPublishDate",
}
```

```python
api = Cov19API(filters=london_only, structure=cases_and_deaths)
data = api.get_dataframe()
```

```python
data.head()
```

```python
food = pd.read_csv("Data/year_osward_grocery.csv")
food.head()
```

```python
mapping = pd.read_csv('Data/Ward_to_Local_Authority_District_(December_2019)_Lookup_in_the_United_Kingdom.csv')
mapping.head()
```

```python
mapping_short = mapping.drop(columns=["WD19NM","LAD19NM","FID"])
```

```python
merged_food = mapping_short.merge(food, left_on="WD19CD", right_on="area_id").drop(columns="area_id")
merged_food.head()
```

```python
merged_total = merged_food.merge(data, left_on="LAD19CD",right_on="areaCode")
merged_total.head()
```
